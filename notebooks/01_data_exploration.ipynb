# Roboflow Dataset Exploration
This notebook explores the downloaded malnutrition detection dataset

```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Explore dataset structure
data_dir = 'data/'
for split in ['train', 'val', 'test']:
    split_path = os.path.join(data_dir, split)
    images_path = os.path.join(split_path, 'images')
    
    if os.path.exists(images_path):
        img_count = len(os.listdir(images_path))
        print(f'{split.upper()}: {img_count} images')
    else:
        # Try class-based structure
        healthy_path = os.path.join(split_path, 'healthy')
        malnourished_path = os.path.join(split_path, 'malnourished')
        
        healthy_count = len(os.listdir(healthy_path)) if os.path.exists(healthy_path) else 0
        malnourished_count = len(os.listdir(malnourished_path)) if os.path.exists(malnourished_path) else 0
        
        print(f'{split.upper()}: {healthy_count} healthy, {malnourished_count} malnourished')
```

```python
# Display sample images
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

for idx, split in enumerate(['train', 'val']):
    images_path = os.path.join(data_dir, split, 'images')
    if os.path.exists(images_path):
        images = [f for f in os.listdir(images_path) if f.endswith(('.jpg', '.png'))]
        if images:
            img = Image.open(os.path.join(images_path, images[0]))
            axes[idx, 0].imshow(img)
            axes[idx, 0].set_title(f'{split.upper()} - Image 1')
            
            if len(images) > 1:
                img2 = Image.open(os.path.join(images_path, images[1]))
                axes[idx, 1].imshow(img2)
                axes[idx, 1].set_title(f'{split.upper()} - Image 2')

plt.tight_layout()
plt.show()
```
