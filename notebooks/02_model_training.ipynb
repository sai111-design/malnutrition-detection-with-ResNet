# Model Training
Train the ResNet-50 + ViT hybrid model

```python
import sys
sys.path.insert(0, '..')

from src.model import MalnutritionDetector
from src.utils import get_device
from src.data_loader import get_roboflow_dataloaders
import torch

device = get_device()
print('Loading data...')
train_loader, val_loader, test_loader = get_roboflow_dataloaders()

print('Creating model...')
model = MalnutritionDetector(pretrained=True).to(device)
```

```python
# Quick test inference
print('Testing model inference...')
model.eval()
with torch.no_grad():
    for images, labels in train_loader:
        images = images.to(device)
        logits = model(images)
        probs = torch.softmax(logits, dim=1)
        print(f'Batch shape: {images.shape}')
        print(f'Output shape: {logits.shape}')
        print(f'Sample predictions: {probs[0]}')
        break
```
